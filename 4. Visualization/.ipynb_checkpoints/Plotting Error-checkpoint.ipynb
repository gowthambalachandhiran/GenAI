```python
import numpy as np
```

```python
X = np.array([-3,-2,-1,1,2,4])
y = np.array([-7,-4,-1,5,8,14])
```

```python
def fun(X,m,c):
    return (m*X)+c
```

```python
print(fun(X,-1,-1))
```

```python
m_values = np.arange(-1, 7)
c_values = np.arange(-1, 3)
```

```python
# Create grid
M, C = np.meshgrid(m_values, c_values)
Z = np.zeros_like(M, dtype=float)
    
```

```python
Z
```

```python
# Calculate sum of squared errors at each (m, c)
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        m = M[i, j]
        c = C[i, j]
        y_pred = fun(X, m, c)
        error = np.sum((y - y_pred) ** 2)
        Z[i, j] = error
```

```python
def sin_func(a,b,X):
    return np.exp(-(a*X))+np.sin(b*X)
```

```python
m_values = np.arange(-3, 7)
c_values = np.arange(-1, 3)
```

```python
# Create grid
M, C = np.meshgrid(m_values, c_values)
Z = np.zeros_like(M, dtype=float)
X_Sin = np.array([1,2,3,4,5,6])
```

```python
y_sin = np.exp(-(3*X_Sin))+np.sin((2*X_Sin))
```

```python
y_sin
```

```python
# Calculate sum of squared errors at each (m, c)
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        m = M[i, j]
        c = C[i, j]
        y_pred = sin_func(m,c,X_Sin)
        error = np.sum((y - y_pred) ** 2)
        Z[i, j] = error
```

```python
# Plotting the surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surf = ax.plot_surface(M, C, Z, cmap='viridis', edgecolor='k', alpha=0.8)

# Labels and title
ax.set_xlabel('Slope (m)')
ax.set_ylabel('Intercept (c)')
ax.set_zlabel('Sum of Squared Errors')
ax.set_title('Error Surface for Sin Regression Parameters')

# Color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='SSE')

plt.show()
```
